## DATAFRAME PRINCIPAL

```python
import pandas as pd
def combine_video_data(video_urls):
    # Crear una lista para almacenar los DataFrames individuales
    dataframes = []

    # Iterar sobre cada URL del vector de videos
    for url in video_urls:
        # Obtener el DataFrame del video actual
        df = utils.yt_scraper(url)
        # Añadirlo a la lista de DataFrames
        dataframes.append(df)
    
    # Concatenar todos los DataFrames en uno solo
    combined_df = pd.concat(dataframes, ignore_index=True)
    
    return combined_df
```

```python
# CARGANDO EL VIDEO DEL QUE QUEREMOS LOS COMENTARIOS
video_urls = [
    'video de youtube.com',

]

df = combine_video_data(video_urls)
df
```
